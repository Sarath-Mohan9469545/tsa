```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_excel("/content/New Microsoft Excel Worksheet.xlsx")
```

```python
print(dataset.dtypes)
```

```python
print(dataset.dtypes)
```

```python
y=dataset[["CONSUMER"]]
```

```python
y
```

```python
x=dataset.drop(['CONSUMER'], axis=1)
```

```python
x
```

```python
print(x)
```

```python
plt.rcParams["figure.figsize"] = (20,10)
plt.plot(x["YEAR"],y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x, y)

# Predicting the Test set results
y_pred = regressor.predict(x)
```

```python
plt.plot(x["YEAR"],y)
plt.plot(x["YEAR"],y_pred)
plt.show()
```
