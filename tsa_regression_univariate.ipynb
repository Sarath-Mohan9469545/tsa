```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("/content/tunnel.csv")
df
```

```python
from datetime import datetime as dt
y=df[["NumVehicles"]]
x=df[["Day"]]
```

```python
x= pd.to_datetime(x.iloc[:, 0], infer_datetime_format=True)
```

```python
x
```

```python
plt.rcParams["figure.figsize"] = (15,5.5)
plt.plot(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x_train= np.array([i for i in range(747)]).reshape(-1,1)
y_train=y.values.reshape(-1,1)
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_train)
```

```python
plt.plot(x,y_pred)
plt.plot(x,y)
```
